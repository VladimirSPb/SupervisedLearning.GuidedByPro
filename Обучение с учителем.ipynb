# Описание проекта

Интернет-магазин «В один клик» заметил снижение покупательской активности постоянных клиентов. Привлечение новых клиентов становится менее эффективным, поэтому компания стремится удержать существующих клиентов с помощью персонализированных предложений.

**Цель проекта:** разработать модель, которая предскажет вероятность снижения покупательской активности клиентов в следующие три месяца, и на основе результатов модели выделить сегменты покупателей для разработки персонализированных предложений.

**Задачи проекта:**

- Разработка модели:

    - Собрать и обработать данные по клиентам, включая признаки коммуникации с клиентом, продуктового поведения, финансового поведения и поведения на сайте.
    - Построить модель, предсказывающую снижение покупательской активности, используя различные алгоритмы и методы оптимизации.
    - Оценить важность признаков и проанализировать результаты модели.
- Сегментация покупателей:

    - Используя результаты модели и данные о прибыльности клиентов, выполнить сегментацию покупателей.
    - Проанализировать выбранный сегмент и предложить меры по увеличению его покупательской активности.

**Данные для работы:**

- market_file.csv: данные о поведении покупателя на сайте, коммуникациях с ним и продуктовом поведении.
- market_money.csv: данные о выручке от покупателя за разные периоды.
- market_time.csv: данные о времени, проведённом на сайте в течение разных периодов.
- money.csv: данные о среднемесячной прибыли от покупателя за последние три месяца.

**Исследование значений в датасетах по формальному признаку**

```python
def df_summary(name: str, df: pd.DataFrame) -> None:
    """
    Функция df_summary принимает на вход str DataFrame и выводит три сводные таблицы с общей информацией:
    
    1. Первая таблица содержит:
       - Первые пять строк датафрейма
    
    2. Вторая таблица содержит:
       - Общее количество строк
       - Общее количество столбцов
       - Количество полных дубликатов строк
    
    3. Третяя таблица содержит информацию о каждом признаке:
       - Название признака
       - Количество пропусков для каждого признака
       - Количество уникальных значений для каждого признака
       - Тип данных каждого признака

    Параметры:
    name (str): Входная строка - название датафрейма;
    df (pd.DataFrame): Входной датафрейм, для которого нужно сформировать сводные таблицы.

    Возвращаемое значение:
    None. Функция выводит таблицы напрямую.
    """ 
    
    # Таблица 1: общее число строк, столбцов и дубликатов
    total_rows = df.shape[0]
    total_columns = df.shape[1]
    total_duplicates = df.duplicated().sum()
    
    summary_1 = pd.DataFrame({
        'Число строк': [total_rows],
        'Число признаков': [total_columns],
        'Число полных дубликатов': [total_duplicates]
    })
    summary_1_t = summary_1.T
    summary_1_t.columns = ['Значение']

    # Таблица 2: информация о каждом признаке
    summary_2 = pd.DataFrame({
        'Признак': df.columns,
        'Число пропусков': df.isnull().sum(),
        'Число уникальных': df.nunique(),
        'Тип': df.dtypes
    }).reset_index(drop=True)
    

    # Вывод таблиц
    print(f'\n\nТаблица: {name}')
    
    print('Обзор данных')
    display(df)
    
    print('', 'Общая информация', sep='\n')
    display(summary_1_t)
    
    print('', 'Информация о признаках',sep='\n')
    display(summary_2)
    
    print('', '', sep='\n')
    
    # Вывод уникальных значений признаков
    print('Уникальные значения и дубликаты признаков', ' ', sep='\n')
    for col in df.columns:
        print(f'Признак {col}')
        print(f'Уникальные значения', df[col].sort_values().unique(), sep='\n')
        print(f'Дубликаты', df[col].duplicated().sum(), '', sep='\n')         
```

# Шаг 3. Объединение таблиц

## 3.1. Объединение таблиц

**Трансформация файла market_money перед объединением**

```python
mm_transformed = (
    market_money
    .pivot_table(index='id', columns='период', values='выручка')
    .rename(columns={'предыдущий_месяц' : 'выручка_предыдущий_месяц', 
                     'препредыдущий_месяц' : 'выручка_препредыдущий_месяц',
                     'текущий_месяц' : 'выручка_текущий_месяц'
                    }
           )
)

mm_transformed = mm_transformed[['выручка_препредыдущий_месяц',
                                 'выручка_предыдущий_месяц',
                                 'выручка_текущий_месяц']]

display(mm_transformed.head(3), mm_transformed.shape)
```

**Трансформация файла market_time перед объединением**

```python
mt_transformed = (
    market_time
    .pivot_table(index='id', columns='период', values='минут')
    .rename(columns={'предыдущий_месяц' : 'время_на_сайте_предыдущий_месяц', 
                     'текущий_месяц' : 'время_на_сайте_текущий_месяц',
                    }
           )
)

display(mt_transformed.head(3), mt_transformed.shape)
```

**Объединение датасетов market_file, market_money, market-time**

```python
df_full = (
    market_file
    .merge(mm_transformed, on='id', how='left')
    .merge(mt_transformed, on='id', how='left')
    .merge(money, on='id', how='left')
)

print(f'Размерность market_file до объединения: {market_file.shape}',
      f'Размерность market_file после объединения: {df_full.shape}', sep='\n')
display(df_full.head(3))
df_full.info(memory_usage='deep')
```

# Шаг 4. Исследовательский анализ данных

**Функция - описательная татистика и графическое представление**

```python
def descriptive_statistics(df: pd.DataFrame) -> None:
    '''
    Функция для вычисления и визуализации описательной статистики для всех числовых и категориальных столбцов датафрейма.

    Параметры:
    df (pd.DataFrame): Входной датафрейм, для которого будет рассчитана и отображена статистика.

    Функция выполняет следующие шаги:
    
    1. Рассчитывает описательную статистику для всех числовых столбцов датафрейма (с помощью метода describe()) и выводит результаты в виде таблицы с округлением до двух знаков.
    
    2. Для каждого числового столбца строит два графика:
        - Boxplot (ящик с усами) для оценки распределения, выбросов и медианы.
        - Гистограмма для оценки распределения значений столбца.
    
    3. Для каждого категориального столбца строится график:
        - Countplot (гистограмма частот) для отображения количества записей для каждой уникальной категории.
    
    Для числовых столбцов синхронизируется ось y между графиками Boxplot и Histogram.
    
    Для категориальных столбцов на график добавляются метки с количеством записей для каждой категории.
    
    Примечание:
    Функция ничего не возвращает, она просто отображает графики и таблицы.
    '''
    
    display(round(df.describe().T,2))
    
    for col in df.select_dtypes(exclude=['object']):
        # Создание фигуры и осей с одинаковыми размерами для двух графиков
        fig, axes = plt.subplots(1, 2)  # 1 ряд, 2 графика

        # Построение boxplot на первом графике
        sns.boxplot(data=df[col], orient='v', ax=axes[0])
        axes[0].set_title(col)
        axes[0].set_title(col)
        axes[0].set_ylabel('значение')

        # Построение histogram на втором графике
        axes[1].hist(x=df[col], bins=30, orientation='horizontal', edgecolor='black')
        axes[1].set_title(col)
        axes[1].set_xlabel('число записей')
        axes[1].grid(zorder=0)        
            
        # Синхронизация оси y для обоих графиков
        axes[0].set_ylim(axes[1].get_ylim())

        # Показ графиков
        plt.tight_layout()
        plt.show()
    
    for col in df.select_dtypes(include=['object']):
        
        #plt.figure(figsize=(10, 5))
        
        sns.countplot(x=df[col], edgecolor='black')
        
        for i, value in enumerate(df[col].value_counts()):
            plt.text(i, value, str(value), ha='center', va='bottom')

        plt.xlabel('')
        plt.ylabel('число записей')
        plt.title(col)
        
        plt.tight_layout()
        plt.show()
```

# Шаг 6. Использование пайплайнов

## 6.1. Пайплайн подготовки данных

**Разделение данных на тренировочную и валидационную выборки**

```python
X = df_ready_for_model.drop('покупательская_активность', axis=1)
y = df_ready_for_model['покупательская_активность']
```

```python
cat_cols = X.select_dtypes(exclude='number').columns
num_cols = X.select_dtypes(include='number').columns
```

```python
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=test_size, 
    stratify=y,
    random_state=seed
)
```

```python
df_train = pd.concat([X_train, y_train], axis=1)
df_test = pd.concat([X_test, y_test], axis=1)
df_test.head(2)
```

**Кодирование целевой переменной - 'покупательская_активность'**

```python
le = LabelEncoder()
y_train_en = le.fit_transform(y_train)
y_test_en = le.transform(y_test)

pd.DataFrame({'original': y_train,
              'encoded': y_train_en}).head(3)
```

**Пайплайн подготовки данных**

**Обработка категориальных признаков**

SimpleImputer + OrdinalEncoder + SimpleImputer

```python
cat_pipe = Pipeline([
        ('simpleimputer_before',
         SimpleImputer(missing_values=np.nan,
                       strategy='constant',
                       fill_value=-1)
        ),
        ('encoder',
         OrdinalEncoder(handle_unknown='use_encoded_value',
                        unknown_value=np.nan)
        ),
        ('simpleimputer_after',
         SimpleImputer(missing_values=np.nan,
                       strategy='constant',
                       fill_value=-1)
        ),
])
```

**Обработка количественных признаков**

SimpleImputer + StandardScaler + SimpleImputer

```python
num_pipe = Pipeline([
        ('simpleimputer_before',
         SimpleImputer(missing_values=np.nan,
                       strategy='constant',
                       fill_value=-1)
        ),
        ('encoder', StandardScaler()
        ),
        ('simpleimputer_after',
         SimpleImputer(missing_values=np.nan,
                       strategy='constant',
                       fill_value=-1)
        ),
])
```

**Пайплайн подготовки данных (категориальные и количесвтенные признаки)**

```python
data_preprocessor = ColumnTransformer(
    [
        ('cat', cat_pipe, cat_cols),
        ('num', num_pipe, num_cols),
    ], 
    remainder = 'passthrough'
)
```

**Пайплайн подготовки данных**

1. Этапы подготовки данных:
    
    - Кодирование категориальных признаков:
        - Использовались OrdinalEncoder (DecisionTreeClassifier) и OneHotEncoder (остальные модели) в процессе отбора лучших гиперпараметров.

    - Масштабирование количественных признаков:
        - Использовались MinMaxScaler, и RobustScaler в процессе гиперпараметрического поиска.
    - Для обработки пропущенных значений применен SimpleImputer с заполнением пропусков константным значением -1.

2. Обработка признаков:

    - Категориальные признаки: SimpleImputer + OrdinalEncoder/ OneHotEncoder + SimpleImputer.
    - Количественные признаки: SimpleImputer + StandardScaler + SimpleImputer.
    
3. Интеграция всех этапов в единый пайплайн: 

    - Обработка категориальных и количественных признаков объединена в общий процесс подготовки данных.

## 6.2. Дамми модель (DummyClassifier)

```python
dummy_model = DummyClassifier(
    strategy='most_frequent',
    random_state=seed
)
dummy_model.fit(X_train, y_train_en)

# предсказание на тестовых данных
dummy_model_preds = dummy_model.predict(X_test) + 1

# оценка качества модели по метрике roc_auc_score
recall = recall_score(y_test_en, dummy_model_preds)
precision = precision_score(y_test_en, dummy_model_preds)
f_score = f1_score(y_test_en, dummy_model_preds)

print('recall =', round(recall, 2))
print('precision =', round(precision, 2))
print('f1_score =', round(f_score, 2))
```

**Дамми модель (DummyClassifier)**

- DummyClassifier обучена с целью предоставления базового уровня качества модели.
- Используемая стратегия: most_frequent.
- Результат: Модель демонстрирует значение f1_score = 0.55, recall = 1, precision = 0.38.

## 6.3. Обучение моделей

**Пайплайн обучения моделей**

```python
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('feature_selection', SelectKBest()),
        ('models', LogisticRegression(random_state=seed))
    ]
)
```

**Сетка гиперпараметров**

```python
# Подготовка параметров
params = {
    'LogisticRegression_l1': {
        'models': [LogisticRegression(class_weight='balanced', penalty='l1', solver='saga', random_state=seed)],
        'models__C': np.arange(0.1, 1.1, 0.1),
        'preprocessor__num': [MinMaxScaler(), RobustScaler()],
        'preprocessor__cat': [OneHotEncoder()],
        #'feature_engineering__degree': range(1, 4),
        'feature_selection__k': range(2, 11),
        'feature_selection__score_func': [f_classif, mutual_info_classif]
    },
    'LogisticRegression_l2': {
        'models': [LogisticRegression(class_weight='balanced', penalty='l2', random_state=seed)],
        'models__C': np.arange(0.1, 1.1, 0.1),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler()],
        'preprocessor__cat': [OneHotEncoder()],
        #'feature_engineering__degree': range(1, 4),
        'feature_selection__k': range(2, 11),
        'feature_selection__score_func': [f_classif, mutual_info_classif]
    },
    'LogisticRegression_elasticnet': {
        'models': [LogisticRegression(class_weight='balanced', penalty='elasticnet', solver='saga', random_state=seed)],
        'models__l1_ratio': np.arange(0.1, 1.1, 0.1),
        'models__C': np.arange(0.1, 1.1, 0.1),
        'preprocessor__num': [MinMaxScaler(), RobustScaler()],
        'preprocessor__cat': [OneHotEncoder()],
        #'feature_engineering__degree': range(1, 4),
        'feature_selection__k': range(2, 11),
        'feature_selection__score_func': [f_classif, mutual_info_classif]
    },
    'SVC': {
        'models': [SVC(class_weight='balanced', random_state=seed)],
        'models__C': np.arange(0.1, 1.1, 0.1),
        'models__kernel': ['rbf', 'poly', 'sigmoid'],
        'models__gamma': ['scale', 'auto'],
        'preprocessor__num': [MinMaxScaler(), RobustScaler()],
        'preprocessor__cat': [OneHotEncoder()],
        #'feature_engineering__degree': range(1, 4),
        'feature_selection__k': range(2, 11),
        'feature_selection__score_func': [f_classif, mutual_info_classif]
    },
    'KNeighborsClassifier': {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(1, 101),
        'models__metric': ['euclidean', 'manhattan', 'minkowski'],
        'models__weights': ['uniform', 'distance'],
        'preprocessor__num': [MinMaxScaler(), RobustScaler()],
        'preprocessor__cat': [OneHotEncoder()],
        #'feature_engineering__degree': range(1, 4),
        'feature_selection__k': range(2, 11),
        'feature_selection__score_func': [f_classif, mutual_info_classif]
    },
    'DecisionTreeClassifier': {
        'models': [DecisionTreeClassifier(class_weight='balanced', random_state=seed)],
        'models__max_depth': range(1, 21),
        'models__min_samples_split': range(2, 21),
        'models__min_samples_leaf': range(1, 21),
        'models__max_leaf_nodes': range(2, 21),
        'preprocessor__num': [MinMaxScaler(), RobustScaler()],
        'preprocessor__cat': [OrdinalEncoder()],
        #'feature_engineering__degree': range(1, 4),
        'feature_selection__k': range(2, 11),
        'feature_selection__score_func': [f_classif, mutual_info_classif]
    }
}
```

**Метрики**

```python
scoring = {
    'f1': 'f1',
    'roc_auc': 'roc_auc',
    'precision': 'precision',
    'recall': 'recall'
}
```

**Обучение модели**

```python
results = {}

start = time.time()
for key, val in params.items():
    rs = RandomizedSearchCV(
        pipe_final,
        val,
        scoring=scoring,
        cv=5,
        n_iter=10,
        n_jobs=-1,
        refit='f1',
        random_state=seed
    )
    rs.fit(X_train, y_train_en)

    results[key] = rs
    
rs_search_time = time.time() - start
print(rs_search_time)
```

**Обучение моделей**

1. Пайплайн обучения моделей включает:
    - Кодирование категориальных и масштабирование количественных признаков.
    - Отбор наиболее информативных признаков методом SelectKBest.
    - Учет дисбаланса целевых классов параметром class_weight.

2. Оптимизация гиперпараметров:

    - Использован метод RandomizedSearchCV для более быстрого и качественного поиска оптимальных параметров. Сетка гиперпараметров включает в себя:
    - Подготовку данных (MinMaxScaler, RobustScaler, OrdinalEncoder/ OneHotEncoder).
    - Гиперпараметры этапов селекции признаков и обучения моделей (SelectKBest, LogisticRegression, SVC, KNeighborsClassifier, DecisionTreeClassifier).

3. Обученные модели:

    - LogisticRegression;
    - SVC;
    - KNeighborsClassifier;
    - DecisionTreeClassifier.

4. Метрика для оценки обобщающей способности: 
    - основная метрика - f1_score - среднее гармоническое между precision и recall - основание для выбора метрики - соблюдение баланса между способностью идентифицировать положительный класс с достаточной точностью.

**Функция для подсчета метрик и визуализации**

```python
def valuation(model, train, target):  
    y_pred = model.predict(train)
    y_proba = model.predict_proba(train)[:, 1]
    
    fig, ax = plt.subplots(2, 2, figsize=(15, 10))  # 2 строки, 2 столбца графиков
    
    # Матрица ошибок
    ConfusionMatrixDisplay.from_estimator(
        model, train, target, ax=ax[0, 0]
    )
    ax[0, 0].set_title('Confusion Matrix') 
    
    # ROC-кривая
    RocCurveDisplay.from_estimator(
        model, train, target, ax=ax[0, 1]
    )
    ax[0, 1].set_title('ROC Curve')
    ax[0, 1].grid(True)
    
    # Используем precision_recall_curve для вычисления precision, recall и порогов
    precisions, recalls, thresholds = precision_recall_curve(target, y_proba)

    table = pd.DataFrame({
        'precision': precisions[:-1],
        'recall': recalls[:-1],
        'threshold': thresholds
    })
    
    # Линейные графики precision и recall
    sns.lineplot(data=table, x='threshold', y='precision', label='Precision', ax=ax[1, 0])
    sns.lineplot(data=table, x='threshold', y='recall', label='Recall', ax=ax[1, 0])
    ax[1, 0].axvline(x=0.5, color='red', linestyle='--', label='Threshold = 0.5')  # Вертикальная линия
    ax[1, 0].set_title('Precision and Recall vs Threshold')
    ax[1, 0].set_xlabel('Threshold')
    ax[1, 0].set_ylabel('Score')
    ax[1, 0].legend()
    ax[1, 0].grid(True)
    
    # Пустой график или возможность добавить дополнительный
    ax[1, 1].axis('off')

    # Вычисление метрик
    roc_auc = np.round(roc_auc_score(target, y_proba), 2)
    precision = np.round(precision_score(target, y_pred), 2)
    recall = np.round(recall_score(target, y_pred), 2)
    f_score = np.round(f1_score(target, y_pred), 2)
    
    print(
        f'ROC_AUC_score: {roc_auc}',
        f'Precision: {precision}',
        f'Recall: {recall}',
        f'F1_score: {f_score}',
        sep='\n')
    
    plt.tight_layout()  # Устранение наложения графиков
    plt.show()
    
    return
```

**Сборка итогового датасета**

```python
# Функция для подготовки данных к передачи в модель

def get_transformed_data(pipeline, data):
    # Извлекаем шаги из пайплайна
    preprocessor = pipeline.named_steps['preprocessor']
    select_kbest = pipeline.named_steps['feature_selection']
    
    # Поэтапно трансформируем данные
    data_preprocessed = preprocessor.transform(data)
    data_selected = select_kbest.transform(data_preprocessed)
    
    return data_selected

# Получение преобразованных данных
data_transformed = get_transformed_data(best_estimator, X_train)
```

```python
# Ручной расчет вероятностей отнесения к классу (leave-one-out)

best_model = best_estimator.named_steps['models']

def leave_one_out_predict_proba(model, X_train, y_train):
    probabilities = []
    distances = pairwise_distances(X_train)
    
    for i, dist in enumerate(distances):
        # Сортируем индексы соседей, исключая текущий объект
        neighbors = np.argsort(dist)[1:model.n_neighbors+1]
        neighbor_classes = y_train[neighbors]
        
        # Рассчитываем вероятности для классов
        prob = np.bincount(neighbor_classes, minlength=len(np.unique(y_train))) / model.n_neighbors
        probabilities.append(prob)
    
    return np.array(probabilities)

# Веротяности отнесения к классу тренировочных данных
proba_train = leave_one_out_predict_proba(best_model, data_transformed, y_train_en)
```
